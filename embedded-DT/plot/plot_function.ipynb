```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


# Read the data from the files
with open('time.txt', 'r') as f:
    time_data = [float(line.strip()) for line in f.readlines()]

with open('dt.txt', 'r') as f:
    dt_data = [float(line.strip()) for line in f.readlines()]

with open('physical.txt', 'r') as f:
    physical_data = [float(line.strip()) for line in f.readlines()]

# Create a DataFrame from the data
df = pd.DataFrame({'time': time_data, 'dt': dt_data, 'physical': physical_data})

# Create a figure with two subplots, with the second subplot stretched vertically
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 5), gridspec_kw={"height_ratios": [3, 1]})
# Plot the data using seaborn in the first subplot
sns.lineplot(data=df, x='time', y='dt', label='dt', ax=ax1)
sns.lineplot(data=df, x='time', y='physical', label='physical', ax=ax1)
ax1.set_xlabel(' ')
ax1.set_ylabel('Position (degree)')
ax1.legend()
ax1.grid()

# Compute the error and plot it using seaborn in the second subplot
df['error'] = df['physical'] - df['dt']
sns.lineplot(data=df, x='time', y='error', ax=ax2)
ax2.set_xlabel('Time(s)')
ax2.set_ylabel('Error (degree)')
ax2.grid()
plt.savefig('my_figure.pdf')
plt.show()
```
